# Recommendation System

### Build a recommender system by using cosine simillarties score.

```python
import pandas as pd 
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
book = pd.read_csv("C:\\Users\\User\\Documents\\Datascience_Assignments\\Assignment 10 (Recommendation system)\\book.csv", engine='python')
book.head(10)
```

```python
book = book.drop(['Unnamed: 0'], axis = 1)
book.head(10)
```

```python
book.Book_Title.nunique()
```

```python
book.User_ID.nunique()
```

```python
book.Book_Rating.nunique()
```

```python
book_df = book.pivot_table(index = 'User_ID',columns = 'Book_Title',values = 'Book_Rating').reset_index(drop = True)
book_df.head(10)
```

```python
book_df.index = book.User_ID.unique()
book_df.head(10)
```

```python
book_df.fillna(0,inplace = True)
book_df.head(10)
```

```python
books_similar = 1 - pairwise_distances(book_df.values,metric = 'cosine')
books_similar
```

```python
books_similar_df = pd.DataFrame(books_similar)
books_similar_df.head(10)
```

```python
books_similar_df.index = book.User_ID.unique()
books_similar_df.columns = book.User_ID.unique()
books_similar_df.head(10)
```

```python
np.fill_diagonal(books_similar,0)
books_similar_df.head(10)
```

```python
books_similar_df.idxmax(axis = 1)
```

```python
similar_books = book_df["You Don't Need Experience if You've Got Attitude"]
similar_books.head(10)
```

```python
similar_to_book = books_similar_df.corrwith(similar_books)
similar_books.head(10)
```

```python
corr_book = pd.DataFrame(similar_to_book, columns = ['Correlation'])
corr_book.dropna(inplace = True)
corr_book
```
